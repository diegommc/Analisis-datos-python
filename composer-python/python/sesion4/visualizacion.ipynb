```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#grafico linea x y deben  tener misma dimension 
x= [1,2,3,4,5]
y=[3,5,6,7,11]
```

```python
plt.plot(x,y)
plt.xlabel('eje x')
plt.ylabel('Eje y')
plt.title('grafico 1')           
```

```python
#grafico de dispercion  relacion entre 2 variables continuas
plt.scatter(x,y)
```

```python
#histograma
datos = np.random.randn(1000)
```

```python
plt.hist(datos, bins=30, color='skyblue', edgecolor='green')
```

```python
#grafica de barras
categorias= ['x','y','z']
valores=[50,60,70]
```

```python
plt.bar(categorias,valores)
plt.show()#te muestra solo la grafica
```
